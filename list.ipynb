```python
thislist=['aures','kips','don']

print(thislist)
print(thislist[1])

salary= [500,600,700,200,100]

print(len(salary))

print(sum(salary)) 

avarage = (sum(salary))/ (len(salary))
print(avarage)

salary.sort()
print(salary)

salary.sort(reverse=True)
print(salary)

salary.append(300)
print(salary)

#inbuilt function to remove an item
salary.pop(2)
print(salary)

#checks if item is in the list 
if 800 in salary:
    print('yess, 300 iko kwa list')
else:
    print('haiko kwa list')

#change item value
city=['Nairobi','Eldoret','kisumu','Mombasa']
city[1]= 'Meru'
print(city)

#change a range of items
city[1:3]= 'Meru','Samburu'
print(city)

#to insert a value
city.insert(3,'Kilifi')
print(city)

#to extend a list
towns=['lodwar','kajiado','solai']
city.extend(towns)
print(city)

#remove list item
city.remove('solai')
print(city)

#for...loop
for x in city:
    print(x)
```
